```python
import numpy as np
import ej23a as ea
import ej23b as eb

A = np.array([[ 2, -1,  0],
              [-1,  2, -1],
              [ 0, -1,  2]])

B = np.array([[4,  1,  1,  1],
              [1,  3, -1,  1],
              [1, -1,  2,  0],
              [1,  1,  0,  2]])

C = np.array([[ 4,  1, -1,  0],
              [ 1,  3, -1,  0],
              [-1, -1,  5,  2],
              [ 0,  0,  2,  4]])

D = np.array([[ 6,  2,  1, -1],
              [ 2,  4,  1,  0],
              [ 1,  1,  4, -1],
              [-1,  0, -1,  3]])


# A, a)
print("a)", "matriz A:")
for row in A: 
    print(*row, sep='\t')
L = ea.Cholesky(A)
print("L=")
for row in L: print(*row, sep='\t')
print()

print("b)", "matriz A:")
L, U = eb.FactLU(A)
D, U = eb.FactD(U)
print("D=")
for row in D: print(*row, sep='\t')

L, U = eb.CholeskyByD(A)
print("L=")
for row in L: print(*row, sep='\t')
print()



# B, a)
print("a)", "matriz B:")
for row in B: 
    print(*row, sep='\t')
L = ea.Cholesky(B)
print("L=")
for row in L: print(*row, sep='\t')
print()

print("b)", "matriz B:")
L, U = eb.FactLU(B)
D, U = eb.FactD(U)
print("D=")
for row in D: print(*row, sep='\t')

L, U = eb.CholeskyByD(B)
print("L=")
for row in L: print(*row, sep='\t')
print()

# C, a)
print("a)", "matriz C:")
for row in C: 
    print(*row, sep='\t')
L = ea.Cholesky(C)
print("L=")
for row in L: print(*row, sep='\t')
print()

print("b)", "matriz C:")
L, U = eb.FactLU(C)
D, U = eb.FactD(U)
print("D=")
for row in D: print(*row, sep='\t')

L, U = eb.CholeskyByD(C)
print("L=")
for row in L: print(*row, sep='\t')
print()

# D, a)
print("a)", "matriz D:")
for row in D: 
    print(*row, sep='\t')
L = ea.Cholesky(D)
print("L=")
for row in L: print(*row, sep='\t')
print()

print("b)", "matriz D:")
L, U = eb.FactLU(D)
D, U = eb.FactD(U)
print("D=")
for row in D: print(*row, sep='\t')

L, U = eb.CholeskyByD(D)
print("L=")
for row in L: print(*row, sep='\t')
print()

```
