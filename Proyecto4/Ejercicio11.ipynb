Ejercicio 11: Uso de metodo secante y Newton

```python
import matplotlib.pyplot as plt

def secante(f, x1, x2, tol=1e-8):
    """Devuelve la el punto para el cual la función f(x0)=0.
    """
    k = 0
    while abs(x2 - x1) > tol and k < 1000:
        if f(x1) - f(x2) != 0:
            x3 = x2 - f(x1)*(x2 - x1)/(f(x2) - f(x1))
            x1 = x2 
            x2 = x3
        else:
            break
        k = k + 1
    return x3

def ddx(f, x):
    """Calcula la derivada de f en x.
    """
    h = 10e-8
    return (f(x+h) - f(x))/h

def Newton(f, x1, tol=1e-8):
    """En cuentra el f(x0) con el método de Newton.
    """
    k = 0
    x0 = x1+1
    while abs(x1 - x0) > tol and k < 1000:
        if ddx(f, x1) != 0:
            x0 = x1
            x1 = x1 - f(x1)/ddx(f, x1)
        k += 1
    return x1

def f(x): 
    return x**3

P = lambda x: 230*x**4 + 18*x**3 + 9*x**2 - 221*x - 9
x0 = [x/100 for x in range(-100, 100, 1)]
y1 = [P(x) for x in x0]
plt.plot(x0, y1)
# plt.show()

# a) Método de Secante
print("Secante")
x1 = secante(P, -.95, -0.001, 1e-6)
print("x1: {:.6f},\tf(x1): {:.6f}".format(x1, f(x1)))
x2 = secante(P, .13, 1, 1e-6)
print("x2: {:.6f},\tf(x2): {:.6f}".format(x2, f(x2)))


# b) Método de Newton
print("\nNewton")
x1 = Newton(P, -0.45, 1e-6)
print("x1: {:.6f},\tf(x1): {:.6f}".format(x1, f(x1)))
x2 = Newton(P, 0.9623, 1e-6)
print("x2: {:.8f},\tf(x2): {:.6f}".format(x2, f(x2)))

```
