Ejercicio 1 $\\$
Sea $X_1,...,X_n$ una muestra aleatoria de distribucion $F(x)$. Se define la funcion de distribucion empirica como:
$$
F_n(x)=\frac{1}{n} \sum^n_{i=1} \mathbb{I}_{X_i\leq x}, x \in \mathbb{R}
$$
Para $x \in \mathbb{R}$ fijo, mostrar que $\mathbb{E}[F_n(x)]=F(x)$ y $Var[F_n(x)]=\frac{F(x)(1-F(x))}{n}$ 

Solucion 1 $\\$
La primera aclaracion que debemos de hacer es que la funcion idicadora sigue una distribucion bernoulli $\ \mathbb{I}_{X_i \leq x} \sim Bernoulli (p=F(x))$


Ya que $P(\mathbb{I}_{X_i\leq x})$=1 =P $(X_i\leq x)=F(x)$ $\\$
Por lo que $\mathbb{E}[\mathbb{I}_{\{X_i\leq x\}}]=F(x)$ 


Ya visto lo anterior podemos partir con la demostracion:
$$
\mathbb{E}[F_n(x)]=\mathbb{E}[\frac{1}{n}\sum^n_{i=1}\mathbb{I}_{\{X_i\leq x\}}]=\frac{1}{n}\sum^n_{i=1}\mathbb{E}[\mathbb{I}_{\{X_i\leq x\}}]=\frac{1}{n}\sum^n_{i=1}F(x)=\frac{1}{n}*nF(x)=F(x)
$$
Por lo que queda demostrado que $\mathbb{E}[F_n(x)]=F(x)$

Ahora buscamos demostrar que $Var[F_n(x)]=\frac{F(x)(1-F(x))}{n}$ $\\$
Partiremos de la definicion de varianza $Var(X)=\mathbb{E}[X]^2-(\mathbb{E}[X])^2$ y del resultado anterior $\mathbb{E}[F_n(x)]=F(x)$


Entonces tendremos que $Var(\mathbb{I}_{\{X_i\leq x\}})=F(x)-(F(x))^2=F(x)[1-F(x)]$ $\\$


Entonces $Var(F_n(x))=\frac{1}{n^2}\sum^n_{i=1}Var[\mathbb{I}_{\{X_i\leq x\}}]=\frac{1}{n^2}\sum^n_{i=1}F(x)[1-F(x)]=\frac{1}{n^2}*nF(x)[1-F(x)]=\frac{F(x)[1-F(x)]}{n}$ $\\$
Por lo que queda demostrado que $Var[F_n(x)]=\frac{F(x)(1-F(x))}{n}$

Ejercico 2: $\\$
Mostrar que el estadistico $X^2$ de la prueba $\chi^2$ de pearson se puede escribir como:
$$
X^2=\sum^c_{k=1}\frac{o^2_k}{e_k}-n
$$
donde n es el tamaño de muestra c es el numero de categorias, $o_k$ y $e_k$ son las frecuencias observadas y esperadas en al k-esima categoria, respectivamente.

Comenzamos escribiendo el estadistico de al prueba $\chi^2$: 
$$
T=\sum^c_{k=1}\frac{(o_k-e_k)^2}{e_i}
$$

Desarrollamos:
$$
X^2=\sum^c_{k=1}\frac{(o_k-e_k)^2}{e_i}=\frac{o^2_k-2o_ke_k+e^2_k}{e_k}
$$

Separamos terminos
$$
\frac{o^2_k}{e_k}-2o_k+e_k
$$

Aplicamos la suma
$$
X^2=\sum^c_{k=1}\frac{o^2_k}{e_k}-2\sum^c_{k=1}o_k+\sum^c_{k=1}e_k
$$

Como sabemos que $\sum o_k=n$ ya que $o_k$ es la suma de las categorias de las frecuencias esperadas y que $\sum e_k=n$ ya que $e_k=np_k$
$$
X^2=\sum^c_{k=1}\frac{o^2_k}{e_k}-2n+n
$$

Lo que nos deja con:
$$
X^2=\sum^c_{k=1}\frac{o^2_k}{e_k}-n
$$

Ejercico 8 $\\$
En un proceso de frabricacion se seleccionan 100 piezas al azar para medir el numero de defectos en cada una de ellas. Las frecuencias observadas de defectos por pieza son las siguientes:

\begin{array}{|c|c|c|c|c|c|c|c|}
\hline
\text{numero de defectos} & 0 & 1 & 2 & 3 & 4 & 5 & 6 & \text{7 o mas} \\ \hline
\text{frecuencia} & 18 & 23 & 27 & 14 & 9 & 6 & 2 & 1 \\ \hline
\end{array}



Prueba la hipotesis nula de que el numero de defectos sigue una distribucion de poisson. Utiliza un nivel de significancia del 5%

Planteamiento $\\$
$H_0$: Numero de defectos ~ Poisson vs $H_a$: Numero de defectos $\nsim$ Poisson y n=100

Lo primero que haremos sera estimar el parametro $\lambda$ con la media muestral entonces:
$$
\sum kf_k = 0(18)+1(23)+2(27)+3(14)+4(9)+5(6)+6(2)+7(1) = 204
$$
Por lo que $\hat{\lambda} = \frac{204}{100} = 2.04$

Ahora calcularemos las probabilidades teoricas para cada numero de eventos bajo la distribucion poisson con la siguiente formula:
$$
P(X=x_i)=e^{-\lambda}\frac{\lambda^{x_i}}{x_i!}
$$
Entonces $e^{2.04}=0.13$

\begin{array}{|c|c|c|}
\hline
k & \text{Prob Esperada} & \text{Frec. esperada}  \\ \hline
0 & 0.13 & 13  \\ \hline
1 & 0.265 & 26.5 \\ \hline
2 & 0.271 & 27.1 \\ \hline
3 & 0.184 & 18.4 \\ \hline
4 & 0.094 & 9.4 \\ \hline
5 & 0.038 & 3.8 \\ \hline
6 & 0.013 & 1.3 \\ \hline
7 & 0.005 & 0.5 \\ \hline
\end{array}

Pero como las utlimas 3 tienen frecuencia esperada <5 las combinaremos 

\begin{array}{|c|c|c|}
\hline
k & \text{Prob Esperada} & \text{Frec. esperada}  \\ \hline
0 & 0.13 & 13  \\ \hline
1 & 0.265 & 26.5 \\ \hline
2 & 0.271 & 27.1 \\ \hline
3 & 0.184 & 18.4 \\ \hline
4 & 0.094 & 9.4 \\ \hline
5 & - & 5.6 \\ \hline
\end{array}

Y aplicando la formula del estadistico de prueba lo calculamos:
$$
\chi^2=\frac{(18-13)^2}{13}+\frac{(23-26.5)^2}{26.5}+\frac{(27-27.1)^2}{27.1}+\frac{(14-18.4)^2}{18.4}+\frac{(9-9.4)^2}{9.4}+\frac{(9-5.6)^2}{5.6}
$$

Por lo que $\chi^2=5.48$ y como lo grados de libertad son Cateforias - 1 - parametros estimados = 4 $\\$
Entonces el valor critico es $\chi^2_{0.05,4}=9.49$

Como $5.48<9.49$ no rechazamos la hipotesis nula por lo que concluimos que con un nivel de significancia del 5% no hay evidencia suficiente para decir que el numero de defectos no sigue una distribucion poisson. 

Ejercicio practico $\\$
Usted invierte dinero en la compra de una acción. Para saber qué tan viable es su inversión,
ha decidido hacer un análisis de serie de tiempo, pero antes necesita saber si los precios son
aleatorio o no. Este tipo de análisis se utiliza para examinar los datos históricos y predecir
futuros cambios en el valor de una inversión. Antes de invertir, es importante informarse y
evaluar los rendimientos

```python
import pandas as pd
import numpy as np
import math
from scipy import stats
import matplotlib.pyplot as plt


spy = pd.read_csv("spy.csv")


spy = spy.sort_values("Date")


spy["Return"] = spy["Close"].pct_change()


x = spy["Return"].dropna()
x = np.sort(x)


spy_ret = spy["Close"].pct_change()


spy_ret = spy.dropna()

 
def empirical(data):
    n = len(data)
    muestra = np.sort(data)
    empirical_values = []
    for x in muestra: 
        count = muestra <= x
        empirical_values.append( np.sum(count) / n)
    return empirical_values

F_n = empirical(x)
# Calculo de los parametros 
n = len(x)
mu = np.mean(x)
sigma2 = np.var(x, ddof=1)
sigma = np.sqrt(sigma2)

F_x = stats.norm.cdf(x, loc=mu, scale=sigma)

#Se cosntruye el dataframe
lillie = pd.DataFrame({'x': x,'F(x)': F_x, 'Fn(i)': F_n})

#La desfasamos
lillie['Fn(i-1)'] = lillie['Fn(i)'] - 1 / n

#D+ y D-
lillie['D+'] = abs(lillie['F(x)']-lillie['Fn(i)'])
lillie['D-'] = abs(lillie['F(x)']-lillie['Fn(i-1)'])

D_mas = lillie['D+'].max()
D_menos = lillie['D-'].max()

Dn = max(D_mas, D_menos)

#El nivel de significancia es del 5%
alpha = 0.05
w_alpha = 0.875897/np.sqrt(n)

#Decision
if Dn > w_alpha:
    print('Rechaza H0: los datos no siguen una distribucion normal')
else:
    print('No rechaza H0: los datos siguen una distribucion normal')
```
